```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("dataset_labels.csv")
```

```python
label_counts = df['label'].value_counts()
```

```python
plt.figure(figsize=(10, 6))
label_counts.plot(kind='bar', color='red')
```

```python
plt.figure(figsize=(8, 8))
label_counts.plot(kind='pie', autopct='%1.1f%%', startangle=140, textprops={'fontsize': 12})
```
